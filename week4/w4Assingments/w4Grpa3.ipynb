```python
# A function to find the longest path in a directed acyclic graph
def longestPath(graph, start, visited, path):
  # Mark the start node as visited and add it to the path
  visited.add(start)
  path.append(start)
  # Initialize the maximum length and the best path
  maxLength = 0
  bestPath = path.copy()
  # Loop through the adjacent nodes of the start node
  for node in graph[start]:
    # If the node is not visited, recursively find the longest path from it
    if node not in visited:
      length, newPath = longestPath(graph, node, visited, path)
      # If the length is greater than the maximum length, update the maximum length and the best path
      if length > maxLength:
        maxLength = length
        bestPath = newPath
  # Remove the start node from the path and the visited set
  path.pop()
  visited.remove(start)
  # Return the maximum length and the best path
  return maxLength + 1, bestPath

# A function to find the longest journey from north to south
def longJourney(AList):
  # Initialize an empty set to store the visited nodes
  visited = set()
  # Initialize an empty list to store the current path
  path = []
  # Initialize the maximum length and the best path
  maxLength = 0
  bestPath = []
  # Loop through the nodes in the graph
  for node in AList:
    # If the node is not visited, find the longest path from it
    if node not in visited:
      length, newPath = longestPath(AList, node, visited, path)
      # If the length is greater than the maximum length, update the maximum length and the best path
      if length > maxLength:
        maxLength = length
        bestPath = newPath
  # Return the best path
  return bestPath
```

This function uses a recursive approach to find the longest path in a directed acyclic graph, where the graph represents the cities and the routes between them. The function also assumes that the graph does not contain any cycles, which means that there is no way to travel back to the north.

The function works by looping through the nodes in the graph and finding the longest path from each node, using a helper function longestPath(). The helper function marks the start node as visited and adds it to the path, then loops through the adjacent nodes of the start node and recursively finds the longest path from each of them. The helper function returns the maximum length and the best path from the start node, and updates the global maximum length and the best path accordingly.
