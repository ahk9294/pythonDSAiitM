```python
def mergeInPlace(A, B):
    # Initialize the indices for A and B
    i = 0
    j = 0
    # Loop until one of the lists is exhausted
    while i < len(A) and j < len(B):
        # If the value at A[i] is greater than the value at B[j]
        if A[i] > B[j]:
            # Swap the values
            A.swap(i, B, j)
            # Increment the index for A
            i += 1
            # Find the correct position for the swapped value in B
            # Initialize a variable to store the current index in B
            k = j
            # Loop until the end of B or until the value at B[k] is less than or equal to the value at B[k+1]
            while k < len(B) - 1 and B[k] > B[k+1]:
                # Swap the values
                B.swap(k, B, k+1)
                # Increment the index for B
                k += 1
        # Otherwise, if the value at A[i] is less than or equal to the value at B[j]
        else:
            # Increment the index for A
            i += 1
    # Return nothing as the lists are merged in place
    return


```
