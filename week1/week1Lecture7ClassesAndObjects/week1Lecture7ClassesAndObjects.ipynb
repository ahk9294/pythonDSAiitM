## Building an intuition about the philosophy of Class and Objects in programming-

<hr>

Consider organisms like Sparrow, Peacock, Hen, Penguin, Ostrich, Koel, Crow.... All of them are like ***objects*** of ***same type*** or ***class*** called "Birds". As they have some ***same*** general structure and ***properties***.

The ***Class*** defines the rules about how the Objects belonging to them 'looks' and 'behave'.

More specifically in programming a ***Class*** is "the place" where *properties* and *functions* for the ***Objects*** belonging to them are defined or written. Its roughly like the **constitution** for the Objects. Defining what "actions" are *legal* or  *illegal* for the Objects belonging to that Class.

In Python a **Class** is defined as follows:

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def walk(self):
        print("I can walk on four legs!")
```
The *nuances* of this code will be discussed later in this Lecture. For Different programming languages we have different way of defining the ***class***. For example the above class in Javascript can be written as :

```javascript
class Animal {
    constructor(name) {
        this.name = name;
    }

    walk() {
        console.log("I can walk on four legs!");
    }
}
```

Once a ***Class*** is created or loaded with all the **characterstics** and **functions** we can create as many **Objects** out of it as we want as follows :

In python object creation from The **Class** is done by following syntax:

```python
bull1=Animal('heera')
bull2=Animal('Moti')
dog= Animal('Sheru')
```

In Javascript the syntax for object creation is :
```javascript
var bull1= new Animal('heera')
var bull2= new Animal('Moti')
var dog= new Animal('Sheru')
```

In the world of computer programming : The objects **bull1** , **bull2**,  and **Dog** created from the same class **Animal** are also called by the term ***instances*** of that ***Class***.

Now the most important point to catch is that - 

-  Every 'instance' of the sames 'Class' shares or "***inherits***" all the properties and behaviour defined in that class.

   - Thus the Objects created for **Animal** class - `bull1` , `bull2`, `dog` - 'shares' or ***inherits*** the `walk` property written in the Animal class.

   - The property of **inheritance** is a very-very important feature of any "Programming language".

   Infact it is not wrong to say that whole **feature of Class-Object Model** was developed into the programming language in order to ultimatly achevie this **inheritance property**.


## why the class-object kind of things are needed?

<hr>

There are situations during programming when we want to create some custom kinds of *objects* which are of 'same type' and have their own special behaviour, due to which we require some new types of operators to work with them. These operators may not be pre-build within the Python.

Consider a situation for example, when  we want to work with ***Cartesian-coordinate Points*** . These **Points** are structured as *tuples of two, three or..more numbers* and when we want to add them they are added in a special way. This feature/notion or **datatype** of a **Cartesian coordinate point** is  not *pre-build* in Python like `integers`.. or `strings`...or `dictionary`.... 

Thus , if we want to work with ***cartesian-coordinate points***  in the context of a programming language we need to 'implement' or 'build' it inside that programming language in the **custom way**.... Now this is when Class- Object feature of a programming language is brought into the scene :

   - First we make a **Class** name **Point**, where we define rules about **how a Point "*looks like*"** and **how to add them**....
   
   - Once this Class **Point** is created we can create as many Point **Objects** as we want out of it and  work with them with our custom defined rules.

We will build this class later in this lecture, at this point we must have a 'broad - understanding' of what this Class-Object Model is all about.

The programming language which allows us to create custom ***Classes and Objects*** in their code, falls in a special category called ***Object-Oriented Programming Languages*** or (OOPLs).

It must be Noted here that the "implementation" of this **Class- Object Model**  in different OOPLs is different.

- For example in **Python** , **Java**, **C++** and **C#**... the ***Objects*** cannot be created without explicitly defining the ***Classes*** first. -- These are called - **Class - Based OOPls**

- There are other OOPLs which treat ***Objects***  as stand-alone separate ***datatype***.In these `objects` can be created as simply as declaring an`int`,`boolean` or `string`. Means we don't need an explicit **Class** to be defined for their creation. 

   - **Javascript** is one of the famous OOPL that falls in this category. 

Now a Valid question at this point is :

 **If Objects can be created without an explicit class in JS, what should we do if we want to create a Multiple object of the "same type" which share "same functions" and "properties" ?** 
 
  or
  
  in-short **How class Model in implemented inside the Javascript?** 

However we must know the fact that In ECMAScript 2015(ES6) revison of the Javascript standards, the  **class model**  was injected inside the main Javascript. It is because the **class model** was widely used by programmers as it makes the code more *neet* and *human readable* .  

But with the injection of **class model** in 2015 JS doenot takes away the ability to create *standalone* objects which were not possible in other OOPLs like Python.

So the question now is : **How class Model in implemented inside the Javascript before 2015?**

- Before the introduction of `class` syntax in ES6, the Javascript 'implements' the class model by using special kinds of `functions` called **constructor functions** and the **inheritance** is acheived using  something called **Prototype chain**. 

Lets have a breif look how it was done then:

- Suppose we want to create the some `objects` of same type in Javascript,First we need something called the **constructor function**. These are special functions but are specially preconfigured by Javascript engineers to create **object instances**.

 To distinguish them from normal functions an convention is followed : **the First letter of their name must be Capital or UpperCase**

 Lets create a **constructor function :**

 ```javascript
 function Animal(name) {
   this.name=name;
 }
 ```
It seems very similar to **Class declaration** in python.

- Now lets create two objects out of it :

```javascript
 var dog= new Animal("Sheru")
 var bull1= new Animal("Heera")
```

Both objects **dog** and **bull1** are the instances of the **Animal constructor**.

- Now we know that both **dog** and **bull1** objects can walk on 4 legs. So we want to create a method **walk()** which we want to be **inherited** by both the instances of the **Animal constructor** .

   Now to 'implement' the very important property of **inheritance** here comes something called **prototype-objects**. Lets understand this.

   - Whenever `object` are created in Javascript, either *standalone* objects or **object instances** created through a **constructor function**, at the same time a -very-special- ***additional 'internal' object assosiated to it*** is also created - called its **prototype-object**. 
   
   - Properties or Fuctions ( or *methods*) written in this **Prototype objects** are directly **inherited** by the respective `object`, to which it is directly assosiated.

   - When we create a **constructor function** to create object instances a single **prototype-object** is created along with it in the background internally. This single **prototype object** of the constructor  will be **inherited by all the object instances created from that constructor**.

   - Now it is preffered that : If ***we think*** that our objects instances have any *method* or *property* that should be ***shared*** with all other object created from the same constructor, then we must put that *property* or *method* in the ***prototype-object*** of the constructor function.

-  Let us now add the **walk()** *method* into the prototype of **Animal** constructor function

```javascript
Animal.prototype.walk= function(){
   console.log("I can walk on four legs!!")
}
```
   As soon as we do that, all the instances of the **Animal constructor** will *inherit* the **walk()** method. We can now directly call the method on the objects:

   ```javascript
   dog.walk() // output: I can walk on four legs!!
   bull1.walk() //output : I can walk on four legs!!
   ```

Let us put all the code together to see how the **inheritance** was acheive in JS before 2015

```javascript
//create constructor function
function Animal(name) {
   //create properties 
   this.name=name;
}

// adding method to prototype for inheritance
Animal.prototype.walk= function () {
   console.log("I can walk on four legs!!");
}

// create object instances
 var dog= new Animal("Sheru")
 var bull1= new Animal("Heera")

//call the methods directly on objects instances

dog.walk() // output: I can walk on four legs!!
bull1.walk() //output : I can walk on four legs!!

```

This is how inheritance was achived before ES6

Lets now see how things changed after introducton of **class model** in JS.

As javascript accepts the class model the code written above can be re written as :


```javascript

class Animal {

   constructor(name) {
      this.name=name;
   }

   // Methods added to prototype directly
  walk() {
   console.log("I can walk on four legs!!")
  }

}

// lets create object instances

var dog=Animal('Sheru')
var bull1=Animal('Heera')

// calling the methods directly on the objects

dog.walk()    // output: I can walk on four legs.
bull1.walk()  // output: I can walk on four legs.
```

 
 One can see that the introduction of **class model** improve the code *readability*
 
 It is very much similar to the python code.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def walk(self):
        print("I can walk on four legs!")
```

## Recognizing difference between custom created and build in objects

If we closely notice, the philosophy of class and their objects is present in Python.

When we write things like :  **x=12** , **y=1.003** , **s="abcd"** we are creating 'objects' of some 'type' or 'class'.

- **x** after storing 12 in it becomes an 'object' of 'class' Integer  or **int**.

- **y** after storing 1.003 in it becomes an 'object' of 'class' **float**.

- **s** is an object of 'class' **string**.

...In this way does this means that every Build-in  Datatype that we study like **list** and **Dictionary** are also 'objects' of some 'class'?

The answer will be Yes! . Every Datatype in Python is an 'Object' of some 'Class' wether its a 'list' , 'Dictionary' or any other. 

Objects belonging to a Class have their **own specific** ***methods*** or  ***functions***.

 For example Objects of Class **Lists** have `.append()` function , which is not available for the Objects of Class `int`.

Similarly Objects of Class **Dictionary** have `**.keys()**` function, which is not available for any Object which belongs to some other Class. 

More intrestingly the behaviour of operators like **+**, **-** are also different for the Objects belonging to different Classes.

If we apply '+' operator on two Objects of **integer** class the result will be the 'arithmetic sum'.                                                                          
But for two objects of **string**' Class the behaviour of same '+' operator is very different. (it **concatinate** two **strings**).

So every Class have their own definitions of operators!!

Now the question is : **"Where" are these Class specific special functions and rules for operators are written?**... 

The answer is - inside the ***Class Definition*** of that Class.

**Where can we find these 'Class Definition'? Say Class definition of 'String' ?**

***Class Definitions*** of things like `string`, `int`,`Dictionary`,`list` etc are written somewhere inside the body of Python code implicitly - that is why these are called ***Built-in Classes*** Or ***Biuld in Datatypes***.

<hr>

Here are some of the main built-in data types in Python:

1. **Numeric Types:**
   - `int`: Integer type, e.g., `42`, `-10`.
   - `float`: Floating-point type, e.g., `3.14`, `-0.5`.
   - `complex`: Complex number type, e.g., `2 + 3j`.

2. **Text Type:**
   - `str`: String type, e.g., `'Hello'`, `"Python"`.

3. **Sequence Types:**
   - `list`: List type, e.g., `[1, 2, 3]`.
   - `tuple`: Tuple type, e.g., `(1, 2, 3)`.

4. **Set Types:**
   - `set`: Set type, e.g., `{1, 2, 3}`.

5. **Mapping Type:**
   - `dict`: Dictionary type, e.g., `{'key': 'value'}`.

6. **Boolean Type:**
   - `bool`: Boolean type, either `True` or `False`.

7. **None Type:**
   - `None`: Represents the absence of a value.

So in Python, there are no **Objects** without a **class**.
