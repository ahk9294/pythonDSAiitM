### Factorial using recursion

Take the base case : $ 1!=1$

````python
def factorial(n):
    #first define the 'base-case'
    if(n==1):
        return(1)
    else:
        return (n*factorial(n-1)) 
        # Here the 'function' is called within the function
        # this is our "Recursion step".
    
#Checking from front-end
    
print(factorial(5))
````

The code seems to be working fine.

### Sum of n positive integers beganning from 1, using Recursion.

Let the base case be :

**Sum of 1 positive integer starting from 1 = 1**.

lets code this:

````python
def sumOfPositiveIntegers(n):
    
    if(n==1): # base-case
        return 1
    else:
        return(n+sumOfPositiveIntegers(n-1)) # Recursion step
    

# Testing from front-end

print(sumOfPositiveIntegers(5))
````

### Power function using Recursion

Suppose take the *base case* be : $ n^1 = n $

Lets now code the function.

````python
def power(n,e):
    if (e==1): # base-case
        return n
    else:
        return(n*power(n,e-1))  # recursion case.
    
# Testing from front-end
    
print(power(2,5))
````

The code seems to be working fine.

### Count Down using Recursion :

## Can we calculate the gcd of two numbers using recursion?

<hr>

To get the answer.. we have to think - does **gcd** have any ***base-case***?

If the answer to this question comes to be **Yes**, then we can definatly solve to this using the *Recursion Technique*.

It turns out that while finding the *gcd*( Greatest Common Divisor or Fcator) of two numbers, say **m** and **n** , if the small number becomes the ***factor of*** (or *completely divides*) the larger number, then the 'gcd' of both number is the small one itself!! -- Thats of ***base-case***!!

Thus now once it is clear that the **gcd problem** can be solved by recursion let us design our Recursion process.

**Step 1: Desing the *'base-case'***.

Base-case for **gcd** of two numbers is :

```python
if( bigger % smaller == 0 ):
    return(smaller)
```

**Step2: Desing the Recursion step.**

Now what will be our Recursion step.? This is our next challenge.

Once we found the **base-case** we are very clear that the Recursion approach can be applied. So our 50% problem is solved. 

Next 50% problem is tricky. In this we have to desing the "Recursion step".

Lets ask : **What actually happens in the recursion step?**

If we go through the examples discussed above we notice that :

- in recursion step we ***modify the input*** in such a   ways that it takes us ***near*** the ***base-case***.

    In other words somehow we 'dig' *deeper* and *deeper* toward the base case.

- After this we supply this ***modified input*** into the main function again to perform the *actual recursion*.

    'Recursion' is nothing but simply  "a function call within the function".


Thus before performing the 'actual recursion' we have to somehow ***Modify the input***. How would we do that in gcd ?

In case of examples discussed above this modified input was easy and self-evident.

But for the **gcd** function this is not that simple and straight forward.

The idea of modification of input for calculating comes from the *pure mathematics*.

It turns out that there are **two ways** in which the input can be modified..

We will now discuss both the ways and then decide which one is more efficient than other.

### Method 1 :

Before writing the actual code let us first see how we arrived at this code.

Suppose we want the gcd of two numbers **m** and **n**. Say **m** is the *bigger* and **n** is the *smaller* one.

Let **d** be the 'greatest number' which divides both **m** and **n** - i.e their **gcd**.

It means **d** is the factor of both **m** and **n**---> then **d** is also the factor or another number "**m-n**". This can be easily reasoned using simple algebra : 

\begin{align*}
m = d \cdot a \\
n = d \cdot b \\
m-n = d \cdot (a-b)
\end{align*}

So now the problem of finding gcd of **m** and **n** can be *reduced* to ---> finding gcd of *smaller* numbers **n** and **m-n**.

Now this *reduction* steps can goes on till we reach the *base-case* - here we got our *reduction step*.

Now let us code this into a function **gcd3**.

````python
def gcd3(m,n):

    # Clarifying the minimum and maximum number.
    (p,q)=(max([m,n]),min([m,n]))

    # the base-case.
    if(p%q==0):
        return(q)
    else:
        return (gcd3(q,p-q)) # the recursion case.
````

Lets test this code from front-end.

````python
print(gcd3(54,53))
print(gcd3(15,30))
print(gcd3(-2,8))
````

The code seems to work fine.

Now let us talk about the efficience of **gcd3**.

**gcd3** is seems to be a *clever* and a **Non-trivial** solution for the *gcd* problem.

No doubt **gcd3** will perform significantly better in situations like **gcd3(10000,800)**. ( In this case the **gcd1** and **gcd2** will take time proportional to the **minimum** number **800**.)

 But if we carefully think : In some worst cases **gcd3** will make the situation *bad* than the **gcd1** and **gcd2**.

Lets see how ?

Consider the situation **gcd3(2,9999)**.

- as 2 doesnt divided 9999, so recursion takes place with **gcd3(2,9997)**.
- again, this will call **gcd3(2,9995)**.
- then this calls **gcd3(2,9993)**.
- ..
- ..
- ..
(this will go on till)
- **gcd3(2,1)** which finally returns the gcd : **1**.

In this case **gcd3** takes about $\approx 5000$ steps.

Thus **gcd3** seems to takes **steps** or **time** proportional to ***maximum***  of the **m** and **n**.

Situations were lot 'better' in **gcd1** and **gcd2** as they will take time proportional to ***minimum number*** of the **m** and **n**.

So **gcd3**, even though a **Non-trivial** solution of the gcd problem , doesnot "guarentees" the efficienty in all cases... as it performs better in some cases and worst in other.

Lets us now modify the **recursive step** the **gcd3** in a different way.

### Method 2

In *method 1* we reduced the problem of **gcd(m,n)** to a smaller problem **gcd(n,m-n)**. Thus we reduced the problem by *subtraction* in the ***recursive step***.

Let us now design the ***recursive step*** in another way. In this we will *reduce* the problem is some other *fashion* (other than *subtraction*).

We want the gcd of **m** and **n**.

Let **m** is greater than **n**. And suppose **d** is the gcd of both number. 

Now if we ***divide*** the larger **m** by smaller **n** then by  **Euclid Divison formula** we can write.

\begin{align*}
m=n \cdot q + r
\end{align*}

If **d** is the **gcd** of **m** and **n**. So we can always write:

\begin{align*}
m=d \cdot a \\
n=d \cdot b
\end{align*}

lets substitute this in the *divison formula*:

\begin{align*}
d \cdot a = (d \cdot b) \times p + r
\end{align*}

Now it is not difficult to imagine that if **left hand side is the factor of d** then the **right hand side will also be the factor of d** , this indicates **r**  must also has **d** as its factor.$ r= d \cdot c$

Now, the above equation can be written as :

\begin{align*}
d \cdot a = (d \cdot b) \times p + (d \cdot c)
\end{align*} 

Thus we got an *another way* to ***reduce*** the gcd problem. --> If '**d**' is the gcd of **m** and **n** then it is also the factor of the **remainder** that we obtain by dividing **m** by **n**.

This fact came out of the *Euclid division Formula*. So notice how *mathematics* plays a **key role** in solving a problem through programming in **efficient way**.

Now let us modify the **gcd3** code and make a new function **gcd3Mod1** .

````python
def gcd3Mod1(m,n):

    #clarify the minimum and maximum
    (mx,mn)=(max([m,n]),min([m,n]))

    # base-case
    if(mx%mn==0):
        return(mn)
    else:
        return(gcd3Mod1(mn,mx%mn)) # Note the recursive step
    
````

Lets test this code from front-end:

````python
print(gcd3Mod1(54,53))
print(gcd3Mod1(15,30))
print(gcd3Mod1(-2,8))
````

The code seems to be working fine.

Let us now discuss the efficiency of **gcd3Mod1**

<hr>

<sup>[ Prof. Madhvan ]</sup>

We can actually show that the time taken by **gcd3Mod1** takes the time proportional to the *number of digits* in the larger number not to the 'size' of the number.

So in case of gcd(9999,2) the **gcd3Mod1** will take time proportional to '4' as there are 4 digits in 9999 so nearly 4 division takes place till we reach the base-case.

    A 'reasonalble-algorithm' for any 'numerical' problem like this should "*grow*" as the ***number of digits***.

Take for example the problem of 'multiplication' of two numbers : If we do $35 \times 5$ and then we do $354 \times 5$ ,in the 2nd case: the number to be multipled grows 10 times but the only *extra work* that we have to do is to multiply one 'more' *digit*.

    Think how difficult it would become if we need to take *10 times more steps* to multiple if the number only grows by 10 times.

So in the numerical world we really want things which are proportional to the *number of digits*.
  
It can be shown (but we will not proove here) that **Euclid Division Algorithm** to find the gcd have this property to grow with number of 'digits'.

Thus at the end we found a reasonable solution for the gcd problem.

This style of calculating the **gcd** by using the **Euclid Division Formula** is  called **Euclid Divison Algorithm**.

**Euclid Divison Algorithm** is one of the early examples of **Non-Trivial** solution of any problem where we use a completely different approach to solve the gcd problem which is not 'obvious'.

That is why this algorithm is interesting.
